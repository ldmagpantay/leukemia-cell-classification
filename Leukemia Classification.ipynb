```python
# -*- coding: utf-8 -*-

# Form implementation generated from reading ui file 'untitled.ui'
#
# Created by: PyQt5 UI code generator 5.15.4
#
# WARNING: Any manual changes made to this file will be lost when pyuic5 is
# run again.  Do not edit this file unless you know what you are doing.

from PyQt5 import QtCore, QtGui, QtWidgets
from PyQt5.QtWidgets import QApplication, QMainWindow, QDesktopWidget, QFileDialog, QMessageBox, QGridLayout, QPushButton, QSizePolicy
from imageai.Detection.Custom import CustomObjectDetection as COD
from imageai.Detection.Custom import CustomVideoObjectDetection as CVOD
import os
import cv2
import sys
import csv 
import time
import csv
import argparse 
import datetime 
import pandas as pd

#Front-end code

class Ui_MainWindow(object):
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.setEnabled(True)
        MainWindow.resize(908, 881)
        sizePolicy = QtWidgets.QSizePolicy(QtWidgets.QSizePolicy.Fixed, QtWidgets.QSizePolicy.Fixed)
        MainWindow.setMinimumSize(QtCore.QSize(908, 881))
        MainWindow.setMaximumSize(QtCore.QSize(908, 881))
        MainWindow.setCursor(QtGui.QCursor(QtCore.Qt.ArrowCursor))
        MainWindow.setMouseTracking(True)
        
        #Window icons
        icon = QtGui.QIcon()
        icon.addPixmap(QtGui.QPixmap(":/icons/microscope.png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        MainWindow.setWindowIcon(icon)
        MainWindow.setStyleSheet("background-color: pink")
        MainWindow.setTabShape(QtWidgets.QTabWidget.Rounded)
        
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setObjectName("centralwidget")
          
        #App Title
        self.AppTitle = QtWidgets.QLabel(self.centralwidget)
        self.AppTitle.setGeometry(QtCore.QRect(10, -10, 861, 131))
        font = QtGui.QFont()
        font.setFamily("Arial")
        font.setPointSize(72)
        self.AppTitle.setFont(font)
        self.AppTitle.setStyleSheet("")
        self.AppTitle.setFrameShape(QtWidgets.QFrame.NoFrame)
        self.AppTitle.setObjectName("AppTitle")
        
        #Use camera icon
        self.use_camera = QtWidgets.QLabel(self.centralwidget)
        self.use_camera.setGeometry(QtCore.QRect(330, 470, 251, 261))
        self.use_camera.setObjectName("use_camera")
        
        #Get video file icon
        self.get_video = QtWidgets.QLabel(self.centralwidget)
        self.get_video.setGeometry(QtCore.QRect(560, 130, 251, 261))
        self.get_video.setObjectName("get_video")
        
        #Get image file icon
        self.get_image = QtWidgets.QLabel(self.centralwidget)
        self.get_image.setGeometry(QtCore.QRect(70, 130, 261, 261))
        self.get_image.setObjectName("get_image")
        
        #Get image button
        self.get_img_button = QtWidgets.QPushButton(self.centralwidget)
        self.get_img_button.setGeometry(QtCore.QRect(120, 400, 151, 41))
        font = QtGui.QFont()
        font.setFamily("Arial")
        font.setPointSize(16)
        self.get_img_button.setFont(font)
        self.get_img_button.setCursor(QtGui.QCursor(QtCore.Qt.PointingHandCursor))
        self.get_img_button.setAutoFillBackground(False)
        self.get_img_button.setStyleSheet("background-color: rgb(255, 255, 255);")
        self.get_img_button.setInputMethodHints(QtCore.Qt.ImhNone)
        self.get_img_button.setIconSize(QtCore.QSize(20, 20))
        self.get_img_button.setCheckable(False)
        self.get_img_button.setObjectName("get_img_button")
        
        #Use camera button
        self.use_cam_button = QtWidgets.QPushButton(self.centralwidget)
        self.use_cam_button.setGeometry(QtCore.QRect(400, 740, 131, 41))
        font = QtGui.QFont()
        font.setFamily("Arial")
        font.setPointSize(16)
        self.use_cam_button.setFont(font)
        self.use_cam_button.setCursor(QtGui.QCursor(QtCore.Qt.PointingHandCursor))
        self.use_cam_button.setAutoFillBackground(False)
        self.use_cam_button.setStyleSheet("background-color: rgb(255, 255, 255);")
        self.use_cam_button.setInputMethodHints(QtCore.Qt.ImhNone)
        self.use_cam_button.setIconSize(QtCore.QSize(20, 20))
        self.use_cam_button.setCheckable(False)
        self.use_cam_button.setAutoDefault(False)
        self.use_cam_button.setObjectName("use_cam_button")
        
        #Get video file button
        self.get_vid_button = QtWidgets.QPushButton(self.centralwidget)
        self.get_vid_button.setGeometry(QtCore.QRect(610, 400, 151, 41))
        font = QtGui.QFont()
        font.setFamily("Arial")
        font.setPointSize(16)
        self.get_vid_button.setFont(font)
        self.get_vid_button.setCursor(QtGui.QCursor(QtCore.Qt.PointingHandCursor))
        self.get_vid_button.setAutoFillBackground(False)
        self.get_vid_button.setStyleSheet("background-color: rgb(255, 255, 255);")
        self.get_vid_button.setInputMethodHints(QtCore.Qt.ImhNone)
        self.get_vid_button.setIconSize(QtCore.QSize(20, 20))
        self.get_vid_button.setCheckable(False)
        self.get_vid_button.setAutoDefault(False)
        self.get_vid_button.setObjectName("get_vid_button")
        
        #---
        self.AppTitle.raise_()
        self.use_camera.raise_()
        self.get_img_button.raise_()
        self.get_video.raise_()
        self.get_image.raise_()
        self.use_cam_button.raise_()
        self.get_vid_button.raise_()
        MainWindow.setCentralWidget(self.centralwidget)

        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)
        
        
        #Click events
        self.get_img_button.clicked.connect(self.getImage)
        self.get_vid_button.clicked.connect(self.getVideo)
        self.use_cam_button.clicked.connect(self.liveFeed)
#-------------------------------------------------------------------------------------------------------------------------

#Back-end Code

#For getting image file
    #------------------------------------------------------
    def getImage(self):
        fname = QFileDialog.getOpenFileName(MainWindow, 'Open file', 'c:/',"Image files (*.jpg *.png)")
        #Setting Variable name "imagePath" to global so it can be called at def DetectImage()
        global imagePath
        imagePath = fname[0]
        #call function DetectImage()
        self.DetectImage()
        
    def DetectImage(self):
        from datetime import datetime
        today = datetime.now()
        
        #Setting model and paths
        img_detector = COD() #custom obj detection
        img_detector.setModelTypeAsYOLOv3()
        img_detector.setModelPath("./model/detection_model-ex-028--loss-0002.803.h5")
        img_detector.setJsonPath("./json config/detection_config.json")
        img_detector.loadModel()
        detections = img_detector.detectObjectsFromImage(input_image=imagePath, 
                                                     output_image_path= imagePath + "_detected.jpg", 
                                                     minimum_percentage_probability=30)
        #Call function popupImage
        self.popupImage()
        
        #Create CSV File from the image
        with open('single-image-detection.csv', 'w', newline='') as f:
            writer = csv.writer(f)
            #column names
            writer.writerow(['CLASS', 'PERCENTAGE'])
            for eachItem in detections:
                writer.writerow([eachItem["name"], eachItem["percentage_probability"]])
 
    def popupImage(self):
        popupMsg = QMessageBox()
        popupImageIcon = QtGui.QIcon()
        popupMsg.setWindowTitle("Alert")
        popupMsg.setText("Single Image Detection Finished!")
        popupMsg.setIcon(QMessageBox.Information)
        popupImageIcon.addPixmap(QtGui.QPixmap(":/icons/microscope.png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        popupMsg.setWindowIcon(popupImageIcon)
        x = popupMsg.exec_()
    #--------------------------------------------------------------------

#For getting the video file
    #--------------------------------------------------------------------
    def getVideo(self):
        vidname = QFileDialog.getOpenFileName(MainWindow, 'Open file', 'c:/',"Video files (*.avi *.mp4)")
        #Setting Variable name "aviPath" to global so it can be called at def DetectVideo()
        global aviPath
        aviPath = vidname[0]
        #call function DetectVideo
        self.DetectVideo()
    
    def DetectVideo(self):
        from datetime import datetime
        today = datetime.now()
        classid = []
        percentage =[]
        execution_path = os.getcwd()
        
        #For video file detection
        def forFrame(frame_number, output_array, output_count):
            for eachObject in output_array:
                classid.append(eachObject["name"])
                percentage.append(eachObject["percentage_probability"])
                
        #Setting model and paths
        vid_detector = CVOD() #custom video obj detection
        vid_detector.setModelTypeAsYOLOv3()
        vid_detector.setModelPath("./model/detection_model-ex-028--loss-0002.803.h5")
        vid_detector.setJsonPath("./json config/detection_config.json")
        vid_detector.loadModel()
        vid_detector.detectObjectsFromVideo(input_file_path=aviPath,
                                            output_file_path=os.path.join(execution_path, aviPath + " cell_detected"),
                                            per_frame_function=forFrame,
                                            minimum_percentage_probability=30,
                                            log_progress=True) 
        #Call function popupVid()
        self.popupVid()
        
        #Create CSV File from the image
        with open('video-image-detection.csv', 'w', newline='') as f:
            writer = csv.writer(f)
            #column names
            writer.writerow(['CLASS', 'PERCENTAGE'])
            for (classid_x, percentage_x) in zip (classid, percentage):
                writer.writerow([classid_x, percentage_x])
                
    def popupVid(self):
        popupMsg2 = QMessageBox()
        popupImageIcon = QtGui.QIcon()
        popupMsg2.setWindowTitle("Alert")
        popupMsg2.setText("Video Detection Finished!")
        popupMsg2.setIcon(QMessageBox.Information)
        popupImageIcon.addPixmap(QtGui.QPixmap(":/icons/microscope.png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        popupMsg2.setWindowIcon(popupImageIcon) 
        y = popupMsg2.exec_()
    #----------------------------------------------------------------------

#For live feed
    #----------------------------------------------------------------------
    def liveFeed(self):
        from datetime import datetime
        classid = []
        percentage =[]
        execution_path = os.getcwd()
        
        camera = cv2.VideoCapture(0)
        
        #For live feed detection
        def forFrame(frame_number, output_array, output_count):
            for eachObject in output_array:
                classid.append(eachObject["name"])
                percentage.append(eachObject["percentage_probability"])
                
        #Setting model and paths
        live_feed_detector = CVOD() #custom video obj detection
        live_feed_detector.setModelTypeAsYOLOv3()
        live_feed_detector.setModelPath("./model/detection_model-ex-028--loss-0002.803.h5")
        live_feed_detector.setJsonPath("./json config/detection_config.json")
        live_feed_detector.loadModel()
        
        #Timer is Set to 5 Seconds, to change, change dection_timeout value to desired seconds.
        live_feed_detector.detectObjectsFromVideo(camera_input=camera, 
                                                  output_file_path=os.path.join(execution_path, "Live-Feed-Output"),
                                                  per_frame_function=forFrame,  minimum_percentage_probability=50,
                                                  detection_timeout=5)
        #Call function popupLive
        self.popupLive()
        
        #Create CSV File from the image
        with open('live-feed-detection.csv', 'w', newline='') as f:
            writer = csv.writer(f)
            #columns name
            writer.writerow(['CLASS', 'PERCENTAGE'])
            for (classid_x, percentage_x) in zip (classid, percentage):
                writer.writerow([classid_x, percentage_x])
        

    def popupLive(self):
        popupMsg3 = QMessageBox()
        popupImageIcon = QtGui.QIcon()
        popupMsg3.setWindowTitle("Alert")
        popupMsg3.setText("Live Feed Detection Finished!")
        popupMsg3.setIcon(QMessageBox.Information)
        popupImageIcon.addPixmap(QtGui.QPixmap(":/icons/microscope.png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        popupMsg3.setWindowIcon(popupImageIcon) 
        y = popupMsg3.exec_()
        #----------------------------------------------------------------

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "Leukemia Cell Classification"))
        self.get_img_button.setText(_translate("MainWindow", "Get Image File"))
        self.AppTitle.setWhatsThis(_translate("MainWindow", "<html><head/><body><p>App title</p><p><br/></p></body></html>"))
        self.AppTitle.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\"><span style=\" font-size:24pt; font-weight:600;\">Leukemia Cell Classification Using Deep Learning</span></p></body></html>"))
        self.use_camera.setText(_translate("MainWindow", "<html><head/><body><p><img src=\":/icons/use-camera.png\"/></p></body></html>"))
        self.get_video.setText(_translate("MainWindow", "<html><head/><body><p><img src=\":/icons/get-video.png\"/></p></body></html>"))
        self.get_image.setText(_translate("MainWindow", "<html><head/><body><p><img src=\":/icons/get-img.png\"/></p></body></html>"))
        self.use_cam_button.setText(_translate("MainWindow", "Use Camera"))
        self.get_vid_button.setText(_translate("MainWindow", "Get Video File"))
import resource
if __name__ == "__main__":
    import sys
    app = QtWidgets.QApplication(sys.argv)
    MainWindow = QtWidgets.QMainWindow()
    ui = Ui_MainWindow()
    ui.setupUi(MainWindow)
    MainWindow.show()
    sys.exit(app.exec_())
```
